```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from collections import Counter

# Ensure necessary NLTK data is downloaded
nltk.download('punkt')
nltk.download('stopwords')
```

```python
text = """
My hobby is reading, which I enjoy doing every day. I love reading all kinds of books like storybooks, comics, and 
and educational books. Reading helps me learn new things, improve my language skills, and grow my knowledge.
I have my special corner at home where I sit and enjoy my books.
"""
print(text)
```

```python
words = word_tokenize(text.lower())  # Convert to lowercase for uniformity
print(words)
```

```python
stop_words = set(stopwords.words('english'))
filtered_words = [word for word in words if word.isalpha() and word not in stop_words]
print(filtered_words)
```

```python
word_freq = Counter(filtered_words)
print(word_freq)
```

```python
wordcloud = WordCloud(width=800, height=400, background_color='white').generate_from_frequencies(word_freq)
```

```python
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')  # Hide axes
plt.show()
```
